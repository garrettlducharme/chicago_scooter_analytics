According to the [meteostat documentation](https://api.meteostat.net/#history-hourly), the "time_local" variable refers to the local time (Chicago central time) specified in the api call. So, we can safely drop the "time" column from this data. We'll also drop the completely empty columns seen from the results of info above.

```python
weather_data_df.drop(columns=['time', 'peakgust', 'precipitation_3', 
                              'precipitation_6', 'snowdepth'], inplace=True)
```
